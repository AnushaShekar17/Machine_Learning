```python
import os
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# ==== Input from user (simulate text input for now) ====
user_input = {
    "goal": "lose weight",
    "medical_conditions": ["type 2 diabetes"],
    "fitness_routine": "daily jogging",
    "preferences": ["vegetarian", "low sugar"]
}

# ==== Mock meal plan generation ====
meal_plan = {
    "Breakfast": "Oatmeal with chia seeds, almond milk, and berries",
    "Lunch": "Quinoa salad with chickpeas, cucumber, and olive oil",
    "Snack": "Carrot sticks with hummus",
    "Dinner": "Grilled tofu with steamed broccoli and brown rice"
}

# ==== Mock explanation generation ====
explanations = {
    "Oatmeal": "High in fiber and stabilizes blood sugar levels.",
    "Quinoa Salad": "Protein-rich and supports energy levels post-workout.",
    "Carrot Sticks": "Low sugar, high in beta-carotene.",
    "Tofu Dinner": "Low fat, high protein vegetarian option."
}

# ==== Create image with output ====
output_text = "🧠 Smart AI Nutrition Assistant Output\n\n"
output_text += f"🎯 Goal: {user_input['goal'].title()}\n"
output_text += f"🩺 Conditions: {', '.join(user_input['medical_conditions'])}\n"
output_text += f"🏃 Routine: {user_input['fitness_routine']}\n"
output_text += f"🍽 Preferences: {', '.join(user_input['preferences'])}\n\n"
output_text += "📅 Meal Plan:\n"

for meal, item in meal_plan.items():
    output_text += f"  {meal}: {item}\n"
output_text += "\n💡 Explanations:\n"
for item, why in explanations.items():
    output_text += f"  {item}: {why}\n"

# ==== Save output as image ====
img = Image.new('RGB', (1000, 600), color=(255, 255, 255))
draw = ImageDraw.Draw(img)

try:
    font = ImageFont.truetype("arial.ttf", 20)
except:
    font = ImageFont.load_default()

draw.multiline_text((20, 20), output_text, fill=(0, 0, 0), font=font, spacing=6)

# ==== Save to specified path ====
save_path = r"C:\Users\anush\OneDrive\Desktop\ML ASSIGNMENT2\ai_meal_plan_output.png"
os.makedirs(os.path.dirname(save_path), exist_ok=True)
img.save(save_path)

# ==== Display image inline ====
plt.imshow(img)
plt.axis('off')
plt.title("Generated Meal Plan")
plt.show()
```
